```python
police =pd.read_csv('/content/drive/MyDrive/detroit data/Police_Serviced_911_Calls_805604165310935371.csv')
```

```python
df = police.loc[:,['Code Description','Category','Priority','Call Time','Neighborhood Name','Council District']]
```

```python
by_neighborhood = df.groupby(['Neighborhood Name', 'Priority']).count()
by_neighborhood
```

```python
group1 = ['P', '1','2']
group2 = ['3', '4', '5']
highp = ['P','1','2','3']
```

```python
def p1(Priority):
  if Priority in group1:
    return 2
  elif Priority in group2:
    return 1
  else:
    return 0
df['Pgroup1'] = df['Priority'].apply(p1)
df

```

```python
def p2(Priority):
  if Priority in highp:
    return 1
  else:
    return 0
df['PGr2'] = df['Priority'].apply(p2)
```

```python
df[['Month','Day', 'Year']] = df['Call Time'].str.split('/',expand=True)
df=df.drop(columns=['Call Time'])
```

```python
def getYear(y):
 if y:
  return y[0:4]

df['Year'] = df['Year'].apply(getYear)
df
```
